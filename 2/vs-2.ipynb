```python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},
                   columns = ['key', 'data1', 'data2'])
df
```

```python
df2 = df.groupby('key').agg(['min', 'median', 'max'])
df2
```

```python
df2.columns = [f"{t[0]}_{t[1]}" for t in df2.columns]
df2
```

```python
df.groupby('key').agg({'data1': 'min',
                             'data2': 'max'})
```

```python
decade = 10 * (planets['year'] // 10 )
decade = decade.astype(str) + 's'
decade.name = 'decade'
planets.groupby(['method', decade])['number'].sum().unstack().fillna(0)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.groupby('sex')[['survived']].mean()
```

```python
(titanic
 .groupby(['sex', 'class'],
          observed=True)['survived']
 .mean()
 .unstack())
```

```python
titanic.pivot_table('survived', index='sex', columns='class', observed=True)
```

```python
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class', observed=True)
```

```python
fare = pd.qcut(titanic['fare'], 2)
titanic.pivot_table('survived', ['sex', age], [fare, 'class'],  observed=True)
```
